```python
#Generated by chatGPT
from sklearn.linear_model import Ridge, Lasso, LinearRegression
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Load the diabetes dataset
diabetes = load_diabetes()

# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(diabetes.data, diabetes.target, test_size=0.25, random_state=42)

# Instantiate Ridge, Lasso, and Linear regression models 
ridge_model = Ridge(alpha=0.1) # Set by Bae
lasso_model = Lasso(alpha=0.1) # Set by Bae
linear_model = LinearRegression()

# Fit the models to the training data
ridge_model.fit(X_train, y_train)
lasso_model.fit(X_train, y_train)
linear_model.fit(X_train, y_train)

# Predict on the training and test data
ridge_train_preds = ridge_model.predict(X_train)
ridge_test_preds = ridge_model.predict(X_test)
lasso_train_preds = lasso_model.predict(X_train)
lasso_test_preds = lasso_model.predict(X_test)
linear_train_preds = linear_model.predict(X_train)
linear_test_preds = linear_model.predict(X_test)

# Calculate the mean squared error for each model
ridge_train_mse = mean_squared_error(y_train, ridge_train_preds)
ridge_test_mse = mean_squared_error(y_test, ridge_test_preds)
lasso_train_mse = mean_squared_error(y_train, lasso_train_preds)
lasso_test_mse = mean_squared_error(y_test, lasso_test_preds)
linear_train_mse = mean_squared_error(y_train, linear_train_preds)
linear_test_mse = mean_squared_error(y_test, linear_test_preds)

print("Ridge Regression train MSE: ", ridge_train_mse)
print("Ridge Regression validation MSE: ", ridge_test_mse)
print("Lasso Regression train MSE: ", lasso_train_mse)
print("Lasso Regression validation MSE: ", lasso_test_mse)
print("Linear Regression train MSE: ", linear_train_mse)
print("Linear Regression validation MSE: ", linear_test_mse)
```
